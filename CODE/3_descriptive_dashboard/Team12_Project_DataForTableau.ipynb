## Models Data Clean-up

```python
def labels(row):
  if row['targets'] == 1 and row['pred'] == 1 :
      return 'TP'
  if row['targets'] == 1 and row['pred'] == 0 :
      return 'FN'
  if row['targets'] == 0 and row['pred'] == 1 :
      return 'FP'
  if row['targets'] == 0 and row['pred'] == 0 :
      return 'TN'      
```

```python
X = ['train', 'test']
Y = ['death', 'hosp', 'icu']
files = ['preds', 'targets', 'features']

for y in range(len(Y)):
  for x in range(len(X)):
    df_pred = pd.read_csv ('/content/drive/MyDrive/Final_dfs/'+Y[y]+'_'+X[x]+'_'+files[0]+'.csv')
    df_targets = pd.read_csv ('/content/drive/MyDrive/Final_dfs/'+Y[y]+'_'+X[x]+'_'+files[1]+'.csv')
    df_features = pd.read_csv ('/content/drive/MyDrive/Final_dfs/'+Y[y]+'_'+X[x]+'_'+files[2]+'.csv')
    df_features['month'] = (df_features['month']*11)+1 
    df_features['year'] = (df_features['year']*2)+2020
    df_features['year'] = df_features['year'].astype(int)
    df_features['year'] = df_features['year'].astype(str)
    df_features['month_name'] = df_features['month'].apply(lambda x: calendar.month_abbr[int(math.ceil(x))])
    df_features['mon-year'] = df_features['month_name']+'-'+df_features['year']
    print('Size of df: %s %s' %(Y[y],X[x]))
    print(len(df_features))
    globals()['df_%s_%s' %(Y[y],X[x])] = pd.DataFrame(df_features['mon-year'])
    globals()['df_%s_%s' %(Y[y],X[x])]['targets'] = df_targets
    globals()['df_%s_%s' %(Y[y],X[x])]['pred'] = df_pred
    globals()['df_%s_%s' %(Y[y],X[x])]['label'] = globals()['df_%s_%s' %(Y[y],X[x])].apply (lambda row: labels(row), axis=1)
    globals()['df_%s_%s' %(Y[y],X[x])]['model'] = Y[y]
    globals()['df_%s_%s' %(Y[y],X[x])]['type'] = X[x]
```

```python
def labels(row):
  if row['target'] == 1 and row['pred'] == 1 :
      return 'TP'
  if row['target'] == 1 and row['pred'] == 0 :
      return 'FN'
  if row['target'] == 0 and row['pred'] == 1 :
      return 'FP'
  if row['target'] == 0 and row['pred'] == 0 :
      return 'TN' 

def classes(row,col,label):
  if row[col] == label :
    return 1
  else:
    return 0
```
